1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$

$\textbf{Answer}$

Step 1: Assume, for the sake of contradiction, that $e$ is rational. This means that $e$ can be expressed as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers with no common factors other than 1, and $q \neq 0$.

Step 2: Recall the definition of $e$ as the sum of the infinite series:


$ e = \Sigma_{k=0}^\infty \frac{1}{k!}$
​
 
Step 3: Consider the partial sums of the series. Let $S_n$ denote the $n$th partial sum:

$S_n = \Sigma_{k=0}^\infty \frac{1}{k!}$

Step 4: Recognize that $e$ is defined as the limit of these partial sums as $n$ approaches infinity:

$e = \lim_{n \to \infty} S_n$
 
Step 5: Since $e$ is rational, $S_n$ must also approach a rational limit as $n$ goes to infinity.

Step 6: Notice that each term in the series $\frac{1}{k!}$ is positive. Therefore, each partial sum $S_n$ is increasing.

Step 7: Consider the factorial terms $k!$. For any $k \geq 1$, we have $k! > k$, which implies $\frac{1}{k!} < \frac{1}{k}$.

Step 8: Recognize that the series $\sum_{k=0}^\infty \frac{1}{k}$ diverges (known as the harmonic series).

Step 9: Since $\frac{1}{k!} < \frac{1}{k}$, we can conclude that the series $\sum_{k=0}^\infty \frac{1}{k!}$ converges.

Step 10: If $S_n$ converges to a rational number, it must be bounded.

Step 11: But the convergence of $\sum_{k=0}^\infty \frac{1}{k!}$ contradicts the unboundedness of the harmonic series.

$\textbf{Conclusion}$: Since assuming $e$ is rational leads to a contradiction, we conclude that $e$ is not rational, i.e., $e \notin \mathbb{Q}$.

\end{bmatrix}^\frac{1}{p}

\begin{bmatrix} 
\Sigma_{k=1}^n |a_k|^p 
\end{bmatrix}^ \frac{1}{p}

\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}

$\text{Answer:}$

Step 1: Laws of Exponents (Power raised to a power):

- Let $n = $ any base number, then ${(n^p)}^\frac{1}{p}=n$
- We know that ${(n^p)}^\frac{1}{p}=n$ for any base number $n$, which implies that the $p$-th root of $n^p$ is $n$. Applying this to the sums of $|a_k|^p$ and $|b_k|^p$, we have:

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k|
+
\Sigma_{k=1}^n |b_k|
\end{gather}

Step 2: Property of Summation (Addition)

\begin{gather} 
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k| + |b_k|
\end{gather}

Step 3: Prove by Triangle Inequality

- Disregarding the Sigma Notation we get:
\begin{gather}
|a_k + b_k|
\leq
|a_k| + b_k| 
\end{gather}

- Since, $(a_k, b_k) > 0$, then $(a + b) = a + b$, and $|a| = a_k$, $|b| = b$:
- Substitute:
\begin{gather}
|a + b| = a + b = |a| + |b|
\end{gather}

$\textbf{Conclusion}$: Thus, the inequality is proven.

$\textbf{4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:}$

\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

$\text{Answer:}$

#### Solution:


Step 1: To prove Sedrakayan’s Lemma directly, we need to demonstrate that for all $(u_i, v_i \in \mathbb{R}^+)$, the inequality

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2 \leq \frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

holds true. 

Step 2: The ***Cauchy-Schwarz Inequality*** states that for any real numbers $(a_i)$ and $(b_i)$, 

$$\left(\sum_{i=1}^n a_i b_i\right)^2 \leq \left(\sum_{i=1}^n (a_i)^2\right) \left(\sum_{i=1}^n (b_i)^2\right)$$

Step 3: Let $a_i = \sqrt{u_i}$ and $b_i = \frac{1}{\sqrt{v_i}}$. Then, the Cauchy-Schwarz Inequality becomes

$$\left(\sum_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}}\right)^2 \leq \left(\sum_{i=1}^n u_i\right) \left(\sum_{i=1}^n \frac{1}{v_i}\right)$$

#### Simplify:

Step 4: The *left-hand side* simplifies to 

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2$$

Step 5: The *right-hand side* simplifies to 

$$\frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

Therefore, since the left-hand side is less than or equal to the right-hand side, we have proven Sedrakayan’s Lemma.

$\textbf{2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:}$

\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

To establish Sedrakayan's Lemma when $u_i$ and $v_i$ are square roots of even integers:

Let $u_i = \sqrt{2a_i}$ and $v_i = \sqrt{2b_i}$, where $a_i$ and $b_i$ are positive integers. Our goal is to demonstrate:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{(2b_i)}
$$

$\textbf{Step 1: Expand the terms}$

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} = \frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right) \left( \sum_{i=1}^n \sqrt{2a_i} \right)}{\sum_{i=1}^n \sqrt{2b_i}}
$$

and

$$
\sum_{i=1}^n \frac{(2a_i)}{(2b_i)} = \sum_{i=1}^n \frac{2a_i}{2b_i}
$$

$\textbf{Step 2: Rewrite the inequality}$

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{\sqrt{2a_i}^2}{\sqrt{2b_i}}
$$

$\textbf{Step 3: Apply the identity}$ 

$$
\left( \sum_{i=1}^n x_i \right)^2 = \sum_{i=1}^n x_i^2 + 2 \sum_{1 \leq i < j \leq n} x_i x_j
$$

Applying this to the numerator, we get:

$$
\frac{\left( \sum_{i=1}^n \sqrt{2a_i} \right)^2}{\sum_{i=1}^n \sqrt{2b_i}} = \frac{\sum_{i=1}^n (2a_i) + 2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}}{\sum_{i=1}^n \sqrt{2b_i}}
$$

$\textbf{Step 4: Notice that $2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}$ is even.}$

$\textbf{Step 5: Combine the terms} $

$$
\frac{\sum_{i=1}^n (2a_i)}{\sum_{i=1}^n \sqrt{2b_i}} + \frac{2 \sum_{1 \leq i < j \leq n} \sqrt{2a_i} \sqrt{2a_j}}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{\sqrt{2b_i}}
$$

$\textbf{Step 6: Simplify}$

$$
\frac{\sum_{i=1}^n (2a_i)}{\sum_{i=1}^n \sqrt{2b_i}} \leq \sum_{i=1}^n \frac{(2a_i)}{\sqrt{2b_i}}
$$

Since both sides are non-negative and the numerator of the left-hand side contains the terms of the denominator of the right-hand side, the inequality holds.

$\textbf{Conclusion:}$ Since both sides are non-negative and the numerator of the left-hand side contains the terms of the denominator of the right-hand side, the inequality holds. Thus, Sedrakayan's Lemma is established for $u_i$ and $v_i$ being square roots of even integers.
